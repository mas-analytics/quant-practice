```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# データ取得（Apple）
df = yf.download("AAPL", start="2020-01-01", end="2024-01-01")

# 最初の5行表示
df.head()
```

```python
df["Close"].plot(title="AAPL Price")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# データ取得（Apple）
df = yf.download("NVDA", start="2020-01-01", end="2024-01-01")

# 最初の5行表示
df.head()
```

```python
df["Close"].plot(title="NVDAPrice")
plt.show()
```

```python
# リターン計算（対数リターン）
import numpy as np

df["Return"] = np.log(df["Close"] / df["Close"].shift(1))

df[["Close", "Return"]].head()
```

```python
df["Return"].plot(title="AAPL Log Returns")
plt.show()
```

```python
df["Return"].hist(bins=50)
plt.title("Return Distribution")
plt.show()
```
